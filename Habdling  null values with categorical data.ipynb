```python
c_d['SalePrice'] = df['SalePrice']
```

```python
c_d.head()
```

```python
## Before 

fig = plt.figure(figsize = (6,3))

ax = fig.add_subplot(1,1,1)

c_d[c_d['BsmtQual']== 'TA']['SalePrice'].plot(kind = 'kde',ax = ax, label = 'Normal data',color='y')
c_d[c_d['BsmtQual'].isnull()]['SalePrice'].plot(kind = 'kde',ax = ax, label = 'NULL  obser  data',color='b')
plt.legend(loc =0)
plt.show()
```

```python
## TA observation i am to save compare with filled null values
```

```python
temp = c_d[c_d['BsmtQual']== 'TA']['SalePrice']
```

```python
c_d['BsmtQual'].fillna('TA',inplace = True)
```

```python
fig = plt.figure(figsize = (6,3))

ax = fig.add_subplot(1,1,1)
temp.plot(kind = 'kde', ax = ax, label = 'before data',color = 'r')
## after
c_d[c_d['BsmtQual']== 'TA']['SalePrice'].plot(kind = 'kde',ax = ax, label = 'after replaceing data',color='y')

plt.legend(loc =0)
plt.show()
```
