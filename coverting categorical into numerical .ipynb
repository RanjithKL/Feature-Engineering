## count and frequency encoding

```python
df = pd.read_csv('./house_price.csv')
```

```python
df.head(10)
```

```python
d1 = df[['Neighborhood','Exterior1st']]
```

```python
d1
```

```python
d1['Neighborhood'].value_counts()
```

```python
d1['Neighborhood'].unique()
```

```python
len(d1['Neighborhood'].unique())
```

```python
d1.shape
```

## count Technique

```python
value = d1['Neighborhood'].value_counts().to_dict()
value
```

```python
d1['Neighborhood_replaced'] = d1['Neighborhood'].map(value)
```

```python
d1.head(10)
```

## Frequency Technique

```python
len(d1)
```

```python
value1 = (d1['Neighborhood'].value_counts()/len(d1)).to_dict()
value1
```

```python
d1['Neighborhood_replaced_frequency'] = d1['Neighborhood'].map(value1)
```

```python
d1.head(10)
```

## mean encoding

```python
d1['target'] =df['SalePrice']
```

```python
d1.head()
```

```python
value2 =d1.groupby(['Exterior1st'])['target'].mean().to_dict()
```

```python
value2
```

```python
d1['Exterior1st_mean_replaced'] = d1['Exterior1st'].map(value2)
```

```python
d1.head()
```

## Target Guided encoding

```python
d1.head()
```

```python
value3 =d1.groupby(['Exterior1st'])['target'].mean().to_dict()
```

```python
value3
```

```python
value3 =d1.groupby(['Exterior1st'])['target'].mean().sort_values(ascending = True).index
```

```python
value3
```

```python
value3 =d1.groupby(['Neighborhood'])['target'].mean().sort_values(ascending = True).index
value3
```

```python
sol ={ j:i for i,j in enumerate (value3,0)}
```

```python
sol
```

```python
d1['Neighborhood_target_replaced'] = d1['Neighborhood'].map(sol)
```
